# import CSV file

```python
data = pd.read_csv('train.csv')
```

# Change datatypes 

```python
data.dtypes
```

```python
dt = ['Survived', 'Pclass','Sex','Embarked','Ticket','Cabin']
```

```python
data[dt]=data[dt].astype('category')
```

```python
data.dtypes
```

# Drop unwanted columns

```python
data.drop(['PassengerId','Name'], inplace=True, axis = 1)
```

```python
data.columns
```

# Fill missing values

```python
data['Age']=data['Age'].fillna(data['Age'].mean())
```

```python
data['Age'].describe()
```

# Dummies

```python
sex = pd.get_dummies(data.Sex) 
sex.head()
```

```python
embarked = pd.get_dummies(data.Embarked)
embarked.head()
```

```python
data
```

```python
data=pd.concat([data,sex,embarked], axis =1)
data
```

# drop unwanted columns

Cabin should be drop because more than 50% of the data values are missing.
Ticket should be drop has it is a unique number given to the individual which will not make any difference.

```python
data1 = data.drop(['Sex', 'Ticket', 'Cabin', 'Embarked'], axis = 1)
data1
```

# train_test_split

```python
 x_train,x_test,y_train,y_test= train_test_split(data1.drop(['Survived'], axis = 1),data1['Survived'], test_size=0.2)
```
