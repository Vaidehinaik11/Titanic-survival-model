## Fill missing value

```python
test['Age']=test['Age'].fillna(test['Age'].mean())
```

```python
test['Age'].describe()
```

```python
test.isna().sum()
```

```python
test['Fare']=test['Fare'].fillna(0)
```

## Dummies

```python
sex = pd.get_dummies(test.Sex) 
sex.head()
```

```python
embarked = pd.get_dummies(test.Embarked)
embarked.head()
```

```python
test
```

```python
test=pd.concat([test,sex,embarked], axis =1)
test
```

## Drop unwanted columns

```python
test1 = test.drop(['Name','Sex', 'Ticket', 'Cabin', 'Embarked'], axis = 1)
test1
```
